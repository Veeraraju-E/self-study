```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
import pandas as pd
```

## Line Plot

```python
from matplotlib import style
style.use('ggplot')
x1=[1,8,10]
y1=[12,16,6]
x2=[2,9,11]
y2=[6,15,7]

plt.title('More Style')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.plot(x1,y1,color='green',label='A',linewidth=5) #labelling the curve
plt.plot(x2,y2,color='blue',label='B',linewidth=5)

plt.grid(True, color='black')
plt.xticks(np.arange(0,12))
plt.yticks(np.arange(0,17))
plt.legend(loc=(0,1))

plt.show()
```

## Bar Chart

```python
plt.bar([1,3,5,7,9],[5,2,7,8,2],label='Bar1',color='red',width=0.6) #default width is 0.8
plt.bar([2,4,6,8,10],[8,6,2,5,6],label='Bar2',color='green',width=0.6)

plt.title('Bar Chart')
plt.xlabel('Numerically Coded Categories')
plt.ylabel('Frequencies')

plt.legend()
plt.show()
```

## Histogram

```python
#only for one numerical varible
pop_ages=[22,55,62,45,21,22,34,42,48,4,99,102,110,120,75,55,85,86,13,45,64,48,78,25,14,10]
class_width=range(0,120,10)

plt.hist(pop_ages,class_width,rwidth=0.9,color='blue')
plt.grid(True,color='k')

plt.xlabel('Age')
plt.ylabel('Number of ppl')
plt.title('Number of ppl in given age ditribution')


plt.show()
```

## Scatterplot

```python
x=[1,2,3,4,5,6,7,8]
y=[5,2,4,2,1,4,5,2]

plt.scatter(x,y,color='green',s=50,marker='s',label='Points')

plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Random Scatterplot')

plt.grid(True,color='black')
plt.legend(loc=(1,1))

plt.show()
```

## Stackplot
- aka areaplot

```python
days=[1,2,3,4,5]
sleeping=[8,8,6,10,7]
eating=[2,3,4,2,1]
playing=[5,3,3,4,6]

#trick for the double lines
plt.plot([],color='green',label='sleeping',linewidth=5)
plt.plot([],color='red',label='eating',linewidth=5)
plt.plot([],color='blue',label='playing',linewidth=5)

plt.stackplot(days,sleeping,eating,playing,colors=['green','red','blue'])#make sure it's in a list

plt.xlabel('Daily Activities')
plt.ylabel('Num of hours every day')
plt.title('Time spent on daily activities')
plt.xticks(range(1,6,1))

plt.legend(loc=(1,1))

plt.show()
```

## Pie Chart

```python
label=['a','b','c','d','e']
values=[1,2,3,4,5]
explode_num=[0,0,0,0,0.1]
colors_list=['g','grey','y','r','c']
plt.pie(values,labels=label,autopct="%.1f%%",explode=explode_num,colors=colors_list,startangle=90,shadow=True)

plt.title('Random Pie Chart')
plt.legend()
plt.show()
```

## Boxplot

```python
plt.boxplot([10,11,9,17,15,25,50,25,34,56,10,5,8],vert=False)
plt.grid(True,color='red')

plt.title('Random Boxplot')
plt.xlabel('Numbers')
plt.show()
```
