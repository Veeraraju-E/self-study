### Question 3 Read all product sales data and show it  using a multiline plot

Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).

The graph should look like this.



```python
month=df1['month_number'].tolist() #tolist() is very important
face_cream=df1['facecream'].tolist()
face_wash=df1['facewash'].tolist()
toothpaste=df1['toothpaste'].tolist()
bathingsoap=df1['bathingsoap'].tolist()
shampoo=df1['shampoo'].tolist()
moisturizer=df1['moisturizer'].tolist()

plt.plot(month, face_cream,   label = 'Face cream Sales Data', marker='o', linewidth=3)
plt.plot(month, face_wash,   label = 'Face Wash Sales Data',  marker='o', linewidth=3)
plt.plot(month, toothpaste, label = 'ToothPaste Sales Data', marker='o', linewidth=3)
plt.plot(month, bathingsoap, label = 'ToothPaste Sales Data', marker='o', linewidth=3)
plt.plot(month, shampoo, label = 'ToothPaste Sales Data', marker='o', linewidth=3)
plt.plot(month, moisturizer, label = 'ToothPaste Sales Data', marker='o', linewidth=3)

plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc=(0,1))
plt.xticks(month)
plt.yticks(np.arange(0,19000,2000))
plt.title('Sales data')
plt.show()
```

### Question 4: Read toothpaste sales data of each month and show it using a scatter plot
Also, add a grid in the plot. gridline style should â€œâ€“â€œ.

The scatter plot should look like this.



```python
toothpasteData=df1['toothpaste']

plt.scatter(month,toothpasteData,label = 'Toothpaste Sales Data', marker='o')
plt.xticks(month)
plt.yticks(np.arange(4500,9000,500))
plt.xlabel('Month Number')
plt.ylabel('Number of Units Sold')
plt.title('Toothpaste Sales Data')
plt.grid(True, color='black')
plt.show()
```

### Question 5: Read face cream and facewash product sales data and show it using the bar chart
The bar chart should display the number of units sold per month for each product. 

Add a separate bar for each product in the same chart.

```python
plt.bar(month,[a+0.2 for a in face_cream],label='Face Cream Sales Data',color='blue',width=0.3)###############
plt.bar(month,[a-0.2 for a in face_wash],label='Face Wash Sales Data',color='orange',width=0.3)

plt.xlabel('Month Number')
plt.ylabel('Sales Units in Number')
plt.xticks(month)
plt.title('Face Cream vs Face Wash')
           
plt.show()
```

### Question 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges

The histogram should look like this.



```python
b=df1['total_profit']
plt.hist(b,label='Profit data',color='blue',rwidth=0.8)
plt.xlabel('profit data in dollar')
plt.ylabel('actual profit in dollar')
plt.title('Profit data')

plt.yticks(np.arange(0,6,1))
plt.xticks(np.arange(150000,450000,50000))
plt.legend(loc=(0,1))

plt.show()
```

### Question 8: Calculate total sale data for last year for each product and show it using a Pie chart

Note: In Pie chart display Number of units sold per year for each product in percentage.

The Pie chart should look like this.



```python
bs_sum=df1['bathingsoap'].sum()
fw_sum=df1['facewash'].sum()
tp_sum=df1['toothpaste'].sum()
fc_sum=df1['facewash'].sum()
sh_sum=df1['shampoo'].sum()
mst_sum=df1['moisturizer'].sum()

parts=[bs_sum,fw_sum,tp_sum,fc_sum,sh_sum,mst_sum]
labels_list=['Bathing soap','Facewash','Toothpaste','Face Cream','Shampoo','Moisturizer']

plt.pie(parts,labels=labels_list,autopct="%.1f%%")
plt.title('Sales Data at a glance')
plt.legend(loc=(1,0))

plt.show()
```
