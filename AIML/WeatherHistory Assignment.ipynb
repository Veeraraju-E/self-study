```python
Weather=Weather.drop(['Summary','Daily Summary'],axis=1)
Weather
```

```python
Weather['Precip Type']=Weather['Precip Type'].astype('category')
Weather['Precip Type']=Weather['Precip Type'].cat.codes
Weather
```

```python
Weather.isna().sum()
```

```python
del Weather['Loud Cover']
```

```python
Weather
```

```python
Weather.iloc[:,:6].corr()
```

```python
# we are good to go for IV and DV split
X=Weather.iloc[:,:6]
Y=Weather.iloc[:,6]
X,Y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.9)
```

```python
lin_reg=LinearRegression()
lin_reg.fit(X_train,Y_train)
Y_pred=lin_reg.predict(X_test)
```

```python
r2_score(Y_test,Y_pred)
```
